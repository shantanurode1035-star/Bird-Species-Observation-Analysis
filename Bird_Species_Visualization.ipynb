```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('cleaned_combined.csv')
df['Date'] = pd.to_datetime(df['Date'])

print("Data Loaded!")
print(f"Rows    : {len(df)}")
print(f"Columns : {len(df.columns)}")
```

```python
top15 = df['Common_Name'].value_counts().head(15).reset_index()
top15.columns = ['Species', 'Count']

fig = px.bar(
    top15,
    x='Count',
    y='Species',
    orientation='h',
    title='Top 15 Most Observed Bird Species',
    color='Count',
    color_continuous_scale='Viridis'
)
fig.update_layout(yaxis={'categoryorder': 'total ascending'})
fig.show()
```

```python
habitat = df['Location_Type'].value_counts().reset_index()
habitat.columns = ['Habitat', 'Count']

fig = px.pie(
    habitat,
    values='Count',
    names='Habitat',
    title='Forest vs Grassland Observations',
    color_discrete_sequence=['#2ecc71', '#f39c12'],
    hole=0.4
)
fig.show()
```

```python
park = df.groupby('Admin_Unit_Code').agg(
    Observations   = ('Common_Name', 'count'),
    Unique_Species = ('Common_Name', 'nunique')
).reset_index().sort_values('Unique_Species', ascending=False)

fig = px.bar(
    park,
    x='Admin_Unit_Code',
    y='Unique_Species',
    title='Park-wise Unique Species Count',
    color='Unique_Species',
    color_continuous_scale='Blues',
    text='Unique_Species'
)
fig.show()
```

```python
id_method = df['ID_Method'].value_counts().reset_index()
id_method.columns = ['Method', 'Count']

fig = px.pie(
    id_method,
    values='Count',
    names='Method',
    title='Bird Identification Methods',
    color_discrete_sequence=px.colors.qualitative.Set2,
    hole=0.4
)
fig.show()
```

```python
monthly = df.groupby(['Month', 'Month_Name']).size().reset_index(name='Count')
monthly = monthly.sort_values('Month')

fig = px.bar(
    monthly,
    x='Month_Name',
    y='Count',
    title='Monthly Bird Observations',
    color='Count',
    color_continuous_scale='Oranges',
    text='Count'
)
fig.show()
```

```python
fig = px.histogram(
    df,
    x='Temperature',
    color='Location_Type',
    title='Temperature Distribution by Habitat',
    nbins=40,
    color_discrete_map={'Forest': '#2ecc71', 'Grassland': '#f39c12'},
    opacity=0.7,
    barmode='overlay'
)
fig.show()
```

```python
fig = px.scatter(
    df.sample(2000),
    x='Temperature',
    y='Humidity',
    color='Location_Type',
    title='Temperature vs Humidity by Habitat',
    color_discrete_map={'Forest': '#2ecc71', 'Grassland': '#f39c12'},
    opacity=0.6
)
fig.show()
```

```python
sky = df['Sky'].value_counts().reset_index()
sky.columns = ['Sky', 'Count']

fig = px.bar(
    sky,
    x='Sky',
    y='Count',
    title='Observations by Sky Condition',
    color='Count',
    color_continuous_scale='Blues',
    text='Count'
)
fig.show()
```

```python
sex = df['Sex'].value_counts().reset_index()
sex.columns = ['Sex', 'Count']

fig = px.pie(
    sex,
    values='Count',
    names='Sex',
    title='Sex Distribution of Observed Birds',
    color_discrete_sequence=['#3498db', '#e74c3c', '#95a5a6'],
    hole=0.4
)
fig.show()
```

```python
wl = df[df['PIF_Watchlist_Status'] == True]
wl_counts = wl['Common_Name'].value_counts().reset_index()
wl_counts.columns = ['Species', 'Count']

fig = px.bar(
    wl_counts,
    x='Species',
    y='Count',
    title='Watchlist Species Sightings',
    color='Count',
    color_continuous_scale='Reds',
    text='Count'
)
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
dist = df['Disturbance'].value_counts().reset_index()
dist.columns = ['Disturbance', 'Count']

fig = px.bar(
    dist,
    x='Disturbance',
    y='Count',
    title='Disturbance Effect on Observations',
    color='Count',
    color_continuous_scale='Purples',
    text='Count'
)
fig.update_xaxes(tickangle=20)
fig.show()
```

```python
dist = df['Disturbance'].value_counts().reset_index()
dist.columns = ['Disturbance', 'Count']

fig = px.bar(
    dist,
    x='Disturbance',
    y='Count',
    title='Disturbance Effect on Observations',
    color='Count',
    color_continuous_scale='Purples',
    text='Count'
)
fig.update_xaxes(tickangle=20)
fig.show()
```

```python
hab_species = df.groupby(['Location_Type', 'Common_Name']).size().reset_index(name='Count')
top_species = df['Common_Name'].value_counts().head(15).index.tolist()
hab_filtered = hab_species[hab_species['Common_Name'].isin(top_species)]

fig = px.bar(
    hab_filtered,
    x='Common_Name',
    y='Count',
    color='Location_Type',
    title='Top 15 Species: Forest vs Grassland Comparison',
    color_discrete_map={'Forest': '#2ecc71', 'Grassland': '#f39c12'},
    barmode='group'
)
fig.update_xaxes(tickangle=45)
fig.show()
```
