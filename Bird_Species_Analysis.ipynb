```python
# CELL 2 — Load Data
import pandas as pd
import numpy as np

df = pd.read_csv('cleaned_combined.csv')
df['Date'] = pd.to_datetime(df['Date'])

print("Data Loaded!")
print(f"Rows    : {len(df)}")
print(f"Columns : {len(df.columns)}")
```

```python
# CELL 3 — Species Diversity Metrics
print("=" * 50)
print("SPECIES DIVERSITY METRICS")
print("=" * 50)

print(f"\nTotal Unique Species     : {df['Common_Name'].nunique()}")
print(f"Total Unique Scientific  : {df['Scientific_Name'].nunique()}")

print("\nSpecies Count by Habitat:")
habitat_sp = df.groupby('Location_Type')['Common_Name'].nunique().reset_index()
habitat_sp.columns = ['Habitat','Unique_Species']
print(habitat_sp.to_string(index=False))

forest_sp = set(df[df['Location_Type']=='Forest']['Common_Name'])
grass_sp  = set(df[df['Location_Type']=='Grassland']['Common_Name'])
both_sp   = forest_sp & grass_sp
only_f    = forest_sp - grass_sp
only_g    = grass_sp - forest_sp

print(f"\nSpecies in BOTH habitats : {len(both_sp)}")
print(f"Species ONLY in Forest   : {len(only_f)}")
print(f"Species ONLY in Grassland: {len(only_g)}")

print("\nPark-wise Species Diversity:")
park_sp = df.groupby('Admin_Unit_Code').agg(
    Total_Obs      = ('Common_Name','count'),
    Unique_Species = ('Common_Name','nunique')
).sort_values('Unique_Species', ascending=False)
print(park_sp.to_string())
```

```python
# CELL 4 — Top Species Analysis
print("=" * 50)
print("TOP SPECIES ANALYSIS")
print("=" * 50)

print("\nTop 15 Overall:")
top15 = df['Common_Name'].value_counts().head(15).reset_index()
top15.columns = ['Species','Count']
top15['Percentage'] = (top15['Count']/len(df)*100).round(1)
print(top15.to_string(index=False))

print("\nTop 10 Forest Species:")
top_f = df[df['Location_Type']=='Forest']['Common_Name'].value_counts().head(10).reset_index()
top_f.columns = ['Species','Count']
print(top_f.to_string(index=False))

print("\nTop 10 Grassland Species:")
top_g = df[df['Location_Type']=='Grassland']['Common_Name'].value_counts().head(10).reset_index()
top_g.columns = ['Species','Count']
print(top_g.to_string(index=False))
```

```python
# CELL 5 — Exclusive Species
print("=" * 50)
print("EXCLUSIVE SPECIES")
print("=" * 50)

print("\nForest ONLY Species (Top 10):")
only_f_df = df[df['Common_Name'].isin(only_f)]['Common_Name'].value_counts().head(10).reset_index()
only_f_df.columns = ['Species','Count']
print(only_f_df.to_string(index=False))

print("\nGrassland ONLY Species (Top 10):")
only_g_df = df[df['Common_Name'].isin(only_g)]['Common_Name'].value_counts().head(10).reset_index()
only_g_df.columns = ['Species','Count']
print(only_g_df.to_string(index=False))
```

```python
# CELL 6 — Activity Patterns
print("=" * 50)
print("ACTIVITY PATTERNS")
print("=" * 50)

print("\nID Method (How birds were identified):")
id_m = df['ID_Method'].value_counts().reset_index()
id_m.columns = ['Method','Count']
id_m['Percentage'] = (id_m['Count']/len(df)*100).round(1)
print(id_m.to_string(index=False))

print("\nInterval Length:")
intv = df['Interval_Length'].value_counts().reset_index()
intv.columns = ['Interval','Count']
intv['Percentage'] = (intv['Count']/len(df)*100).round(1)
print(intv.to_string(index=False))

print("\nTop 10 Species by Singing:")
singing = df[df['ID_Method']=='Singing']['Common_Name'].value_counts().head(10).reset_index()
singing.columns = ['Species','Count']
print(singing.to_string(index=False))
```

```python
# CELL 7 — Sex Ratio Analysis
print("=" * 50)
print("SEX RATIO ANALYSIS")
print("=" * 50)

print("\nOverall Sex Distribution:")
sex = df['Sex'].value_counts().reset_index()
sex.columns = ['Sex','Count']
sex['Percentage'] = (sex['Count']/len(df)*100).round(1)
print(sex.to_string(index=False))

print("\nSex Distribution by Habitat:")
sex_hab = df.groupby(['Location_Type','Sex']).size().reset_index(name='Count')
print(sex_hab.to_string(index=False))

print("\nTop 10 Species with Male Sightings:")
male = df[df['Sex']=='Male']['Common_Name'].value_counts().head(10).reset_index()
male.columns = ['Species','Count']
print(male.to_string(index=False))

print("\nTop 10 Species with Female Sightings:")
female = df[df['Sex']=='Female']['Common_Name'].value_counts().head(10).reset_index()
female.columns = ['Species','Count']
print(female.to_string(index=False))
```

```python
# CELL 8 — Distance & Flyover Analysis
print("=" * 50)
print("DISTANCE & FLYOVER ANALYSIS")
print("=" * 50)

print("\nDistance Distribution:")
dist = df['Distance'].value_counts().reset_index()
dist.columns = ['Distance','Count']
dist['Percentage'] = (dist['Count']/len(df)*100).round(1)
print(dist.to_string(index=False))

print("\nTop 10 Species Seen Close (<=50 Meters):")
close = df[df['Distance']=='<= 50 Meters']['Common_Name'].value_counts().head(10).reset_index()
close.columns = ['Species','Count']
print(close.to_string(index=False))

print("\nFlyover Observed:")
fly = df['Flyover_Observed'].value_counts().reset_index()
fly.columns = ['Flyover','Count']
fly['Percentage'] = (fly['Count']/len(df)*100).round(1)
print(fly.to_string(index=False))

print("\nTop 10 Flyover Species:")
flysp = df[df['Flyover_Observed']==True]['Common_Name'].value_counts().head(10).reset_index()
flysp.columns = ['Species','Count']
print(flysp.to_string(index=False))
```

```python
# CELL 9 — Conservation Analysis
print("=" * 50)
print("CONSERVATION ANALYSIS")
print("=" * 50)

wl = df[df['PIF_Watchlist_Status']==True]

print(f"\nTotal Watchlist Sightings : {len(wl):,}")
print(f"% of All Observations    : {len(wl)/len(df)*100:.1f}%")

print("\nWatchlist Species Detail:")
wl_detail = wl.groupby('Common_Name').agg(
    Sightings     = ('Common_Name','count'),
    Parks         = ('Admin_Unit_Code','nunique'),
    Habitat       = ('Location_Type', lambda x: '/'.join(sorted(x.unique()))),
    Avg_Temp      = ('Temperature','mean'),
).round(2).sort_values('Sightings', ascending=False)
print(wl_detail.to_string())

print("\nRegional Stewardship Species (Top 10):")
rs = df[df['Regional_Stewardship_Status']==True]['Common_Name'].value_counts().head(10).reset_index()
rs.columns = ['Species','Count']
print(rs.to_string(index=False))
```

```python
# CELL 10 — Species Analysis Summary
print("=" * 50)
print("SPECIES ANALYSIS - KEY FINDINGS")
print("=" * 50)
print(f"1.  Total unique species         : {df['Common_Name'].nunique()}")
print(f"2.  Most observed species        : {df['Common_Name'].value_counts().index[0]} ({df['Common_Name'].value_counts().iloc[0]:,})")
print(f"3.  Forest only species          : {len(only_f)}")
print(f"4.  Grassland only species       : {len(only_g)}")
print(f"5.  Species in both habitats     : {len(both_sp)}")
print(f"6.  Singing identification       : {(df['ID_Method']=='Singing').sum():,} ({(df['ID_Method']=='Singing').sum()/len(df)*100:.1f}%)")
print(f"7.  Male sightings               : {(df['Sex']=='Male').sum():,} ({(df['Sex']=='Male').sum()/len(df)*100:.1f}%)")
print(f"8.  Female sightings             : {(df['Sex']=='Female').sum():,} ({(df['Sex']=='Female').sum()/len(df)*100:.1f}%)")
print(f"9.  Watchlist species            : {df[df['PIF_Watchlist_Status']==True]['Common_Name'].nunique()}")
print(f"10. Most biodiverse park         : {df.groupby('Admin_Unit_Code')['Common_Name'].nunique().idxmax()}")
print(f"11. Rarest watchlist species     : {df[df['PIF_Watchlist_Status']==True]['Common_Name'].value_counts().index[-1]} ({df[df['PIF_Watchlist_Status']==True]['Common_Name'].value_counts().iloc[-1]} sightings)")
print(f"12. Flyover observations         : {df['Flyover_Observed'].sum():,} ({df['Flyover_Observed'].sum()/len(df)*100:.1f}%)")
```
